# Customers

```python
def generate_customers(n=1000):
    for i in tqdm(range(1,n+1),desc="Inserting Customers"):
        name=faker.name()
        email=faker.unique.email()
        created_at=faker.date_time_between(start_date='-2y',end_date='now')
        cursor.execute("""
            INSERT INTO customer (customer_id, name, email, created_at)
            VALUES (%s, %s, %s, %s)
        """, (i, name,email,created_at))
```

```python
cursor.execute("DELETE FROM customer")
conn.commit()
```

```python
generate_customers(100000)
conn.commit()
```

```python
cursor.execute("SELECT * FROM Customer LIMIT 10")  
rows=cursor.fetchall()
for row in rows:
    print(row)
```

# Products

```python
def generate_products(n=10000):
    product_types=["Smartphone","Laptop","T-Shirt","Backpack","Headphones","Sneakers","Washing Machine","Bluetooth Speaker","Cookware Set",
      "LED TV","Water Bottle","Electric Kettle","Tablet","Gaming Mouse",
        "Camera","Office Chair","Monitor","Fitness Tracker","Air Conditioner",
        "Novel","Board Game","Puzzle","Children's Book"
    ]
    adjectives=["Pro","Max", "Lite","Ultra","Basic","Elite","Smart", "Premium","Advance","Eco"]   
    for i in tqdm(range(1,n+1), desc="Inserting Products"):
        product_name=f"{random.choice(adjectives)} {random.choice(product_types)}"
        price=round(random.uniform(10,2000),2)
        stock=random.randint(10,500)
        cursor.execute("""INSERT INTO product (product_id, name, price, stock)
            VALUES (%s, %s, %s, %s)""", (i,product_name,price,stock))
```

```python
cursor.execute("DELETE FROM product")
conn.commit()
```

```python
generate_products(10000)
conn.commit()
```

```python
cursor.execute("SELECT * FROM product LIMIT 10")  
rows=cursor.fetchall()
for row in rows:
    print(row)
```

# ProductCategory

```python
def generate_product_categories(n_products=10000,n_categories=10):
    for product_id in tqdm(range(1,n_products+1),desc="Inserting ProductCategory"):
        assigned=random.sample(range(1,n_categories+1),random.randint(1,3))
        for category_id in assigned:
            cursor.execute("""INSERT INTO productcategory (product_id,category_id)
                VALUES (%s, %s)""",(product_id, category_id))
```

# OrderItems

```python
def generate_order_items(n_orders=300000, max_products=10000):
    for order_id in tqdm(range(1,n_orders+1), desc="Inserting OrderItems"):
        n_items=random.randint(1,3)  
        product_ids=random.sample(range(1,max_products+1),n_items)
        for product_id in product_ids:
            quantity=random.randint(1,5)
            price=round(random.uniform(10,500),2)
            cursor.execute("""INSERT INTO orderitem (order_id, product_id, quantity, price)
                VALUES (%s, %s, %s, %s)""",(order_id,product_id,quantity,price))
```

# Reviews

```python
def generate_meaningful_review(rating,product_name):
    good_comments=[f"Love this {product_name}, works perfectly!",
        f"Great quality {product_name}, highly recommend.",
        f"This {product_name} exceeded my expectations.",
        f"Very satisfied with the {product_name}.",
        f"Top-notch {product_name}, value for money."
    ]
    bad_comments=[f"Disappointed with this {product_name}.",
        f"The {product_name} stopped working in a week.",
        f"Low quality {product_name}, not worth it.",
        f"Wouldn't recommend this {product_name}.",
        f"Terrible experience with the {product_name}."
    ]
    neutral_comments=[f"The {product_name} is okay - not great, not terrible.",
        f"A decent {product_name} for the price.",
        f"This {product_name} does the job.",
        f"It's just an average {product_name}.",
        f"Expected more from this {product_name}."
    ]
    if rating>=4:
        return random.choice(good_comments)
    elif rating==3:
        return random.choice(neutral_comments)
    else:
        return random.choice(bad_comments)
```

```python
def generate_reviews(n_reviews=100000):
    for review_id in tqdm(range(1,n_reviews+1), desc="Inserting Reviews"):
        customer_id=random.randint(1,100000)
        product_id=random.randint(1,10000)
        rating=random.randint(1,5)
        cursor.execute("SELECT name FROM product WHERE product_id = %s", (product_id,))
        product_name=cursor.fetchone()[0]
        review_text=generate_meaningful_review(rating,product_name)
        review_date=faker.date_time_between(start_date='-1y',end_date='now')
        cursor.execute("""
            INSERT INTO review (review_id, customer_id, product_id, rating, review_text, review_date)
            VALUES (%s, %s, %s, %s, %s, %s)""",(review_id,customer_id,product_id, rating,review_text, review_date))
```

```python
cursor.execute("DELETE FROM review")
conn.commit()
```

```python
generate_reviews()
conn.commit()
print("Review table populated")
```

```python
cursor.execute("SELECT * FROM review LIMIT 10")  
rows=cursor.fetchall()
for row in rows:
    print(row)
```
